* Num ficheiro Python, cria uma aplicação para manipular polinómios com as seguintes operações:

    2. Ler uma lista de polinómios de um ficheiro;

    6. Maior grau: mostra o polinómio de maior grau e indica o seu número de ordem;

```python
p1 = [(1,7), (-3.7,4), (5,3), (-88,0)]
p2 = [(2,2), (17,0)]
p3 = [(7,4), (6,3), (-1,2), (43,0)]
p4 = [(1,4)]
p5 = [(1,2)]
p6 = [(1,3)]
polinomios = [p1,p2,p3,p4,p5,p6]
```

```python
def guardarPolinomios(listap, fnome):
    file = open(fnome,"w")
    for p in listap:
        for termo in p:
            coe, grau = (termo,0)
            file.write(str(coe) + ";" + str(grau) + "#")
        file.write("\n")
    file.close()
guardarPolinomios(polinomios,"polinomios.txt")
```

```python
def derivapolinomio(p):
    derivada = []
    for coeficiente, grau in p :
        if grau > 0 :
            novocoeficiente = grau * coeficiente 
            novograu = grau - 1
            derivada.append((novocoeficiente,novograu))
    return derivada

derivapolinomio(p3)
```

```python
def somapolinomios (p1,p2):
    resultado = []
    i = 0
    j = 0
    while i < len(p1) and j < len(p2):
        (coe1, grau1 ), (coe2, grau2 )= p1[i] , p2[j]
        if grau1 == grau2:
            resultado.append((coe1 + coe2, grau1))
            i += 1
            j += 1
        elif grau1 > grau2:
            resultado.append((coe1, grau1))
            i += 1
        else:
            resultado.append((coe2, grau2))
            j += 1
    while i < len(p1):
        coe, grau = p1[i]
        resultado.append((coe, grau))
        i += 1
    while j < len(p2):
        coe, grau = p2[j]
        resultado.append((coe, grau))
        j += 1

    return resultado

somapolinomios(p1,p2)
```

```python
import matplotlib.pyplot as plt
import numpy as np


polinomios = []

def criapolinomio ():
    coeficientes = []
    grau = int(input("Qual é o grau do polinomio ? "))
    for i in range(grau+1):
        coef = int(input(f"Coeficiente de x^{grau - i}: "))
        coeficientes.append(coef)
    polinomios.append(coeficientes)
    print("Polinomio criado com sucesso!")

def listapolinomios():
    if not polinomios:
        print("Nenhum polinómio disponível.")
        return
    
    print("Lista de polinómios:")
    for i, coeficientes in enumerate(polinomios):
        print(f"Polinómio {i+1}: {coeficientes}")

def calcularPolinomios():
    if not polinomios:
        print("Nenhum polinómio disponível.")
        return
    num_ordem = int(input("Número de ordem do polinómio: "))
    if num_ordem < 0 or num_ordem >= len(polinomios):
        print("Número de ordem inválido.")
        return
    
    x = float(input("Valor de x: "))
    coeficientes = polinomios[num_ordem]
    grau = len(coeficientes) - 1
    resultado = 0
    for i, coef in enumerate(coeficientes):
        resultado += coef * (x ** (grau - i))
    
    print(f"Resultado: {resultado}")

def listapolinomioscomgrau():
    if not polinomios:
        print("Nenhum polinómio disponível.")
        return
    print("Lista de polinómios com grau:")
    for i, coeficientes in enumerate(polinomios):
        grau = len(coeficientes) - 1
        print(f"{i}: {coeficientes} (grau {grau})")

def polinomiodemaiorgrau():
    if not polinomios:
        print("Nenhum polinómio disponível.")
        return
    maior_grau = 0
    indice_maior_grau = 0
    for i, coeficientes in enumerate(polinomios):
        grau = len(coeficientes) - 1
        if grau > maior_grau:
            maior_grau = grau
            indice_maior_grau = i
    print(f"Polinómio com maior grau (número de ordem {indice_maior_grau}): {polinomios[indice_maior_grau]}")

def deriva():
    if not polinomios:
        print("Nenhum polinómio disponível.")
        return
    print("Tabela de derivadas:")
    for i, coeficientes in enumerate(polinomios):
        derivada = [coef * i for i, coef in enumerate(coeficientes)][1:]
        print(f"{i}: {coeficientes} -> {derivada}")

def somarpolinomios():
    if len(polinomios) < 2:
        print("Pelo menos dois polinómios são necessários para a operação de soma.")
        return
    print("Escolha os polinómios para soma:")
    for i, coeficientes in enumerate(polinomios):
        print(f"{i}: {coeficientes}")
    indices = [int(input(f"Número de ordem do {n+1}º polinómio: ")) for n in range(2)]
    for i in indices:
        if i < 0 or i >= len(polinomios):
            print("Número de ordem inválido.")
            return
    
    coeficientes1 = polinomios[indices[0]]
    coeficientes2 = polinomios[indices[1]]
    resultado = [c1 + c2 for c1, c2 in zip(coeficientes1, coeficientes2)]
    polinomios.append(resultado)
    print("Polinómio somado com sucesso!")

def gerargrafico():
    if not polinomios:
        print("Nenhum polinómio disponível.")
        return
    num_ordem = int(input("Número de ordem do polinómio: "))
    if num_ordem < 0 or num_ordem >= len(polinomios):
        print("Número de ordem inválido.")
        return
    coeficientes = polinomios[num_ordem]
    x = np.linspace(-10, 10, 400)
    y = np.polyval(coeficientes, x)
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(f'Gráfico do Polinómio {num_ordem}')
    plt.grid(True)
    plt.show()

def guardar(listap,fnome):
    file = open(fnome,"w")
    for p in listap:
        for termo in p:
            coe, grau = termo
            file.write(str(coe) + ";" + str(grau) + "#")
        file.write("\n")
    file.close()
    print("Polinómios gravados com sucesso!")

def menu():
    menu = ("Menu\nEscolha a opção:\n(1)Criar polinómio\n(2)Listar polinómios\n(3)Calcular valor de um polinómio\n(4)Listar polinómios com grau\n(5)Maior grau\n(6)Derivada\n(7)Somar dois polinómios\n(8)Gerar gráfico de polinómio\n(9)Gravar polinómios em arquivo\n(0)Sair")  
    continuar = True 
    print(menu)
    while continuar :
        menuOption = int(input("Escolha uma funcao:"))
        if menuOption == 1:
            criapolinomio()
        elif menuOption == 2:
            listapolinomios()
        elif menuOption == 3:
            calcularPolinomios()
        elif menuOption == 4:
            listapolinomioscomgrau()
        elif menuOption == 5:
            polinomiodemaiorgrau()
        elif menuOption == 6:
            deriva()
        elif menuOption == 7:
            somapolinomios()
        elif menuOption == 8:
            gerargrafico()
        elif menuOption == 9:
            guardar()
        elif menuOption == 0:
            print("saiu da aplicacao")
            continuar = False

menu()
```
